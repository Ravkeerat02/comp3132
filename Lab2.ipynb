```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Importing the CSV file
DF = pd.read_csv('/content/Wisconsin_Diagnostic_Breast_Cancer.csv')
DF.head()
```

```python
# Printing the shape
print('Shape :', DF.shape)
```

```python
X = DF["Mitoses"].values.reshape(-1, 1)
y = DF["Class"].values.reshape(-1, 1)
```

```python
# Splitting the data into 70% training and 30% training. 
# Setting the randomness involved in the model to 42

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state=25)
X_train
```

```python
# Training the Linear Regression Model
reg = LinearRegression()

# Fitting the line to our data
reg.fit(X_train, y_train)
print(reg.intercept_)
```

```python
# Slope
print(reg.coef_)
```

```python
print(reg.intercept_)
```

```python
# Predicting the crime
score = reg.predict([[3.00]])
score
```
